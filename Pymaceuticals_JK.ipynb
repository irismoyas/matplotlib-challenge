```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data_df=pd.read_csv(mouse_drug_data_to_load)
clinical_data_df=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_df=pd.merge(clinical_data_df,mouse_data_df,on='Mouse ID',how='left')

# Display the data table for preview

combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_df=combined_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)']
avg_tumor_df=grouped_df.mean()
# Convert to DataFrame
avg_tumor_df=avg_tumor_df.reset_index()
# Preview DataFrame
avg_tumor_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_standard_error=grouped_df.sem()
# Convert to DataFrame
tumor_standard_error_df=pd.DataFrame(tumor_standard_error)
# Preview DataFrame
tumor_standard_error_df.reset_index(inplace=True)
tumor_standard_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_avg_tumor_df=avg_tumor_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
# Preview that Reformatting worked
pivot_avg_tumor_df.head()
```

```python
# Generate the Plot (with Error Bars)
drug_list=[('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]
for drug,marker,colors in drug_list:
    ste=tumor_standard_error[drug]
    tumor_treatment_plt=plt.errorbar(pivot_avg_tumor_df.index,pivot_avg_tumor_df[drug],ste,fmt=marker,ls='--',color=colors,label=drug,linewidth=0.5)

# Show the Figure
plt.legend(loc="upper left")
plt.title('Tumor Response To Treatment')
plt.xlabel('Time(Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid()

# Save the Figure
plt.savefig("output/tumor_response_to_treatment.png")
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_grouped_df=combined_df.groupby(['Drug','Timepoint'])['Metastatic Sites']
avg_met_df=met_grouped_df.mean()
# Convert to DataFrame
avg_met_df = avg_met_df.reset_index()
# Preview DataFrame
avg_met_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
ste_met=met_grouped_df.sem()
# Convert to DataFrame
ste_met_df=pd.DataFrame(ste_met)
# Preview DataFrame
ste_met_df.reset_index(inplace=True)
ste_met_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
pivot_avg_met_df=avg_met_df.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')
# Preview that Reformatting worked
pivot_avg_met_df.head()
```

```python
# Generate the Plot (with Error Bars)
drug_list=[('Capomulin','o','red'),('Infubinol','^','blue'),('Ketapril','s','green'),('Placebo','d','black')]
for drug,marker,colors in drug_list:
    ste=ste_met[drug]
    met_treatment_plt=plt.errorbar(pivot_avg_met_df.index,pivot_avg_met_df[drug],ste,fmt=marker,ls='-',color=colors,label=drug,linewidth=0.5)



# Show the Figure
plt.legend(loc="upper left")
plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration(Days)')
plt.ylabel('Met.Sites')
plt.grid()

# Save the Figure
plt.savefig("output/metastatic_spread_during_treatment.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change=(pivot_avg_tumor_df.iloc[-1]/(pivot_avg_tumor_df.iloc[0])-1)*100
# Display the data to confirm
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
drugs_list=['Capomulin','Infubinol','Ketapril','Placebo']
change_list=[(percent_change[drug]) for drug in drugs_list]

# Splice the data between passing and failing drugs
passing=percent_change<0

# Orient widths. Add labels, tick marks, etc. 
change_plt=plt.bar(drugs_list,change_list,width=-1,align="edge",color=passing.map({True:'g',False:'r'}))
plt.grid()
plt.ylim(-30,70)
plt.ylabel('% Tumor Volume Change')
plt.title('Tumor Change Over 45 Day Treatment')
# Use functions to label the percentages of changes
def autolabel(rects):
    for rect in rects:
        height=rect.get_height()
        if height>0:
            label_position=2
        else:
            label_position=-8
        plt.text(rect.get_x() + rect.get_width()/2., label_position,
                '%d' % int(height)+'%',color='white',
                ha='center', va='bottom')
# Call functions to implement the function calls
autolabel(change_plt)

# Save the Figure
plt.savefig(("output/Tumor_Change_Over_45_Day_Treatment.png"))


```
